# Simple Linear Regression

## Importing the libraries

```
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Importing the dataset

```
dataset = pd.read_csv("Salary_Data.csv")
X = dataset.iloc[:,:-1].values #[:,:0] also works
y = dataset.iloc[:,-1].values
```

## Splitting the dataset into the Training set and Test set

```
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
#make random_state = 0 in order to compare well
```

## Training the Simple Linear Regression model on the Training set

```
#import the class needed for simple linear regression, sklearn (linear model)
from sklearn.linear_model import LinearRegression
#create instance or object (ex of name to use, regressor)
slr = LinearRegression() #no internal parameters needed for now
#FIT method is used to train models and Neural Networks or anyting in AI
slr.fit(X_train, y_train)
```

## Predicting the Test set results

```
# regressor.predict(X_test)
#display result in a new frame
y_predicted = slr.predict(X_test)
```

## Visualising the Training set results

```
plt.scatter(X_train, y_train, c = 'red')
#plot regression line
plt.plot(X_train, slr.predict(X_train), c='blue' )#note the x value here
plt.title('Salary vs Experiment[Training set]')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```

## Visualising the Test set results

```
plt.scatter(X_test, y_test, c = 'red')
#plot regression line
plt.plot(X_train, slr.predict(X_train), c='blue' )#No need to replace X_train here
plt.title('Salary vs Experiment[Test set]')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```
