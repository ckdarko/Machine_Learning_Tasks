## Part 2 - Building the ANN

### Initializing the ANN

```
#create a variable that is the neural network itself
#created as an object of a class called Sequential
#other types are Computational Graphs, neurons not connected in successive layers. Ex. Restricted or Deep Boltzmann's Machine
ann = tf.keras.models.Sequential()
```

### Adding the input layer and the first hidden layer

```
#Using the Dense Class
#.add is a method in the Dense Class
#Units = number of hidden neurons 
#how many hidden neurons should I choose? 
#these are hyperparameters which won't be trained during the process. Choose the right number
#of hidden parameters based on Experimentation. Using 6 in this task
#NB::this code is basically for the first hidden layer and not the input layer  
ann.add(tf.keras.layers.Dense(units=6, activation='relu'))
```

### Adding the second hidden layer

```
#same code as the first hidden layer
#you can manipulate the hyperparameters and activation functions and compare accuracy 
ann.add(tf.keras.layers.Dense(units=6, activation='relu'))
```

### Adding the output layer

```
#why didn't we choose 2 units for the output?
#this is because we are predicting a binary variable of 1/0. It has only one dimension
#if we were doing Classification with a non-binary dependent variable (ie dependent variable with 3 classes, we will need 3 dimensions)#HIGHER
#we will need to hot encode such dependent variables too
#better use a SIGMOID activation function for the output layer because it helps predict probability that output outcome is 1.
ann.add(tf.keras.layers.Dense(units=1, activation='sigmoid'))
#when prediction more than two categories, activation should be SoftMax
```

## Part 3 - Training the ANN

### Compiling the ANN

```
#adam is a type of Stochastic Gradient Descent
#for Binary Classification, loss function must be binary_crossentropy
#for Non-Binary Classification, loss must be categorical_crossentropy
ann.compile(optimizer='adam', loss='binary_crossentropy', metrics='accuracy')
```

### Training the ANN on the Training set

```
ann.fit(X_train, y_train, batch_size=32, epochs = 100)
#using Batch Gradient Descent
```
